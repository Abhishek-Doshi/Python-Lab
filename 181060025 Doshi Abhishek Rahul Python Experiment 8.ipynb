```python
import time
import datetime

class ATM_account():
    
    data = {}
    
    def __init__(self, name, mob_no, pin):
        self.name = name
        self.mob_no = mob_no
        self.pin = pin
        self.acc_no = int(time.time())
        self.amount = 0
        self.transactions = []
        self.blocked = False
        ATM_account.data[str(name)] = str(self.acc_no) + ',' + str(self.mob_no) + ',' + str(self.pin)
     
    def extract_data_from_dictionary(self, name):
        acc_no, mob_no, pin = ATM_account.data[str(name)].split(',')
        return int(acc_no), int(mob_no), int(pin), self.amount
         
    def verify_pin(self, name, acc_no, pin):
        if self.blocked == False:
            _acc_no, _mob_no, _pin , _amount = ATM_account.extract_data_from_dictionary(self, name)
            if (_acc_no == acc_no) and (_pin == pin) and (self.pin == pin): return True
            else:
                pin = int(input('Entered pin was incorrect, please enter again:'))
                if (_acc_no == acc_no) and (_pin == pin) and (self.pin == pin): return True
                else:
                    pin = int(input('Incorrect pin, please enter again. Entering the wrong pin will block your account:'))
                    if (_acc_no == acc_no) and (_pin == pin) and (self.pin == pin): return True
        self.blocked = True
        print('Your account is blocked, you need to physically visit the bank.')
        return False

    def withdraw(self):
        amount = float(input('\nEnter amount to withdraw: '))
        pin = int(input('Enter pin:'))
        if self.verify_pin(self.name, self.acc_no, pin):
            if self.amount >= amount:
                self.amount -= amount
                self.transactions.append(['Withdrawn Rs.',amount, datetime.datetime.now()])
                print('Successfully withdrawn Rs.' + str(amount) + ' from your account.')
                print('\nYour bank account has ' + str(self.amount) + ' Rs.\n')
            else: print('Not enough balance.')
    
    def deposit(self):
        amount = float(input('\nEnter amount to deposit: '))
        pin = int(input('Enter pin:'))
        if self.verify_pin(self.name, self.acc_no, pin):
            self.amount += amount
            self.transactions.append(['Deposited Rs.',amount, datetime.datetime.now()])
            print('Successfully deposited Rs.' + str(amount) + ' to your account.')
            print('\nYour bank account has ' + str(self.amount) + ' Rs.\n')
            
    def balance(self):
        pin = int(input('\nEnter pin:'))
        if self.verify_pin(self.name, self.acc_no, pin):
            print('\nYour bank account has ' + str(self.amount) + ' Rs.\n')
    
    def account_information(self):
        pin = int(input('\nEnter pin:'))
        if self.verify_pin(self.name, self.acc_no, pin):
            print('------------------------------------------------------------------------------')
            print("\n\nAccount holder's name: " + str(self.name))
            print("\nAccount holder's mobile number: " + str(self.mob_no) + "\n")
            for transaction in self.transactions:
                print('\n' + str(transaction[0]) + str(transaction[1]) + ' on ' + str(transaction[2]))
            print('\nYour bank account has ' + str(self.amount) + ' Rs.\n')
            print('------------------------------------------------------------------------------')
    
    def pin_change(self):
        pin = int(input('\nEnter previous pin:'))
        if self.verify_pin(self.name, self.acc_no, pin):
            equal_pins = False
            new_pin = pin
            while(~equal_pins):    
                new_pin = int(input('Enter new pin: '))
                ver_pin = int(input('Re-enter new pin to verify: '))
                if new_pin == ver_pin:
                    break
                else: print('Both pins were different.')
                new_pin = pin
            self.pin = new_pin
            ATM_account.data[str(self.name)] = str(self.acc_no) + ',' + str(self.mob_no) + ',' + str(self.pin)
            print('\nPin changed successfully!\n')
```

```python
def create_account(acc_list, acc_no_list):
    name = str(input('\nEnter your name: '))
    mob_no = int(input('Enter mobile number: '))
    equal_pins = False
    while(~equal_pins):    
        pin = int(input('Enter pin: '))
        ver_pin = int(input('Re-enter pin to verify: '))
        if pin == ver_pin:
            break
        else: print('Both pins were different.')
    acc_list.append(ATM_account(name, mob_no, pin))
    acc_no_list.append(acc_list[-1].acc_no)
    print('Account created successfully! Your account number is \033[1m\033[94m' + str(acc_list[-1].acc_no) + '\033[0m\n')
```

```python
def get_account(acc_list, acc_no_list):
    name = str(input("\nEnter account holder's name: "))
    acc_no = int(input('Enter account number: '))
    for idx, _acc_no in  enumerate(acc_no_list):
        if acc_no == _acc_no:
            if name == acc_list[idx].name:
                if acc_list[idx].blocked == False:
                    return acc_list[idx]
                else: 
                    print('Your account is blocked, you need to physically visit the bank.')
                    return None
            else:
                print('Name and account number do not match.')
                return None
    print('No account exists with this account number')
    return None
```

```python
def account_services(account):
    service = int(input('\n 1 for Account Information \n 2 for Balance Enquiry \n 3 for Amount Withdrawal \n 4 for Amount Deposit \n 5 for Pin Change \n\n Enter number of your required service: '))
    if (service == 1): account.account_information()
    elif (service == 2): account.balance()
    elif (service == 3): account.withdraw()
    elif (service == 4): account.deposit()
    elif (service == 5): account.pin_change()
```

```python
def delete_previous_accounts():
    print('\n\033[1m\033[94mNOTE: All the previously created accounts are deleted successfully. Currently there are no accounts in the system.\033[0m\n')
    acc_list = []
    acc_no_list = []
    return acc_list, acc_no_list
```

```python
def initialize_ATM_system(empty_list):
    acc_list, acc_no_list = empty_list
    while(True):
        access = int(input('\nEnter 1 to access your account and 2 to create new account: '))
        if (access == 1):
            account = get_account(acc_list, acc_no_list)
            if (account != None):
                account_services(account)
        elif (access == 2):
            create_account(acc_list, acc_no_list)
        else: print('Invalid key entered.')
```

```python
initialize_ATM_system(delete_previous_accounts())
```
